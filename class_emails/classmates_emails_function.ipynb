```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import requests
```

```python
def classmate_emails(cov_id):

    df = pd.read_csv("sample_student_data.csv")
    df = df.set_index("id_number")
    classes = df.loc[int(cov_id), "classes"].split(",")

    df['classmates'] = df.apply(lambda row: True if any([item in row['classes'] for item in classes]) else False, axis = 1)  
    df['classmates'] = df['classmates']*df['obf_email']
    classmates_emails = df['classmates'].tolist()
    classmates_emails = list(set(classmates_emails))
    classmates_emails.remove('')

    return classmates_emails

classmate_emails("202869")
```
